```python
import pandas as pd         
import seaborn as sns       
import matplotlib.pyplot as plt  
import numpy as np          
from sklearn.preprocessing import StandardScaler

import yfinance as yf


def download_ticker_data(ticker_list, period='22d'):   #22 days for now = monthly 
    """Download historical data for a list of tickers and return dictionary of DataFrames"""
    data = yf.download(
        ticker_list,
        period=period,
        interval='1d',
        group_by='ticker',
        progress=False
    )
    
    # Separate into individual DataFrames
    return {ticker: data[ticker] for ticker in ticker_list if ticker in data.columns.get_level_values(0)}

# Download data for each category
#calling function on each ticker lsit separately 
largecap_data = download_ticker_data(tickers_largecap_luxury)
midcap_data = download_ticker_data(tickers_midcap_luxury)
smallcap_data = download_ticker_data(tickers_smallcap_luxury)

# Example: Access data for Hermès
hermes_df = largecap_data["RMS.PA"]
print(hermes_df.head())
```

```python
#corr within a category -> amongst tickers of same industry large-small-mid any one only 

# Combine 'Close' from all tickers in largecap_data
adj_close_largecap = pd.DataFrame({
    ticker: df['Close'] for ticker, df in largecap_data.items()
})

# Calculate correlation matrix
corr_matrix_largecap = adj_close_largecap.corr()

# Plot heatmap
plt.figure(figsize=(14, 12))
sns.heatmap(corr_matrix_largecap, annot=True, cmap='coolwarm')
plt.title('Large Cap Luxury Brands Correlation Heatmap')
plt.show()
```
