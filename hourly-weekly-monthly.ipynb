```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# --- 1. Ticker List (Representative Tech Tickers) ---
tech_tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'NVDA', 'META', 'TSM', 'AVGO', 'ORCL', 'CSCO', 'INTC', 'QCOM', 'AMD', 'CRM', 'ADBE']
# Let's limit to 10 for quicker processing in this example, you can use the full list
tech_tickers = tech_tickers[:10]

print(f"Selected Tech Tickers: {tech_tickers}")
print("\nIMPORTANT: Using 'Close' prices. These are NOT adjusted for dividends or stock splits.")
print("For analyses sensitive to total return or long-term price continuity, 'Adj Close' is usually preferred.\n")

# --- 2. Data Acquisition ---
# Download data for the last 5-10 years to have enough for yearly trends
try:
    data = yf.download(tech_tickers, start="2014-01-01", end="2024-01-01")
    # *** CHANGED HERE: Using 'Close' instead of 'Adj Close' ***
    close_prices = data['Close'].dropna(how='all') # Drop rows where all tickers are NaN
    # Forward fill to handle missing values for individual stocks
    # then backfill for any leading NaNs if a stock IPO'd later
    close_prices = close_prices.ffill().bfill()
    print("\nDownloaded Close Prices (head):")
    print(close_prices.head())
    if close_prices.isnull().sum().any():
        print("\nWarning: Remaining NaNs after ffill/bfill:")
        print(close_prices.isnull().sum()[close_prices.isnull().sum() > 0])
        close_prices = close_prices.dropna(axis=1, how='any') # Drop any columns that still have NaNs
        print(f"Tickers after dropping columns with NaNs: {close_prices.columns.tolist()}")

except Exception as e:
    print(f"Error downloading data: {e}")
    exit()

if close_prices.empty:
    print("No data to process. Exiting.")
    exit()

# --- 3. Resampling ---
resample_periods = {
    'Daily': 'D',
    'Weekly': 'W-FRI', # End of week on Friday
    'Monthly': 'M',    # End of month
    'Yearly': 'A'     # End of year (or 'Y')
}

resampled_prices = {}
for name, period_code in resample_periods.items():
    if name == 'Daily':
        # *** CHANGED HERE: Using close_prices ***
        resampled_prices[name] = close_prices
    else:
        # *** CHANGED HERE: Using close_prices ***
        resampled_prices[name] = close_prices.resample(period_code).last()

# --- 4. Common Ground / Normalization / Standardization ---

# Method A: Percentage Returns (Primary for "common ground" and statistical comparison)
returns_data = {}
for name, prices_df in resampled_prices.items():
    returns = prices_df.pct_change().dropna()
    returns_data[name] = returns
    print(f"\n{name} Returns (head):")
    print(returns.head())

# Method B: Re-indexing Prices to a Common Start (e.g., 100) for Visual Comparison
normalized_prices_to_100 = {}
for name, prices_df in resampled_prices.items():
    if not prices_df.empty and not prices_df.iloc[0].isnull().all():
         normalized_prices_to_100[name] = (prices_df / prices_df.iloc[0] * 100).dropna(how='all')
    else:
        print(f"Skipping normalization to 100 for {name} due to empty or all-NaN first row.")
        normalized_prices_to_100[name] = pd.DataFrame()

# Method C: Z-score Standardization (on returns, as prices are non-stationary)
standardized_returns_data = {}
scaler = StandardScaler()
for name, returns_df in returns_data.items():
    if not returns_df.empty:
        standardized_returns = pd.DataFrame(scaler.fit_transform(returns_df),
                                            index=returns_df.index,
                                            columns=returns_df.columns)
        standardized_returns_data[name] = standardized_returns
    else:
        standardized_returns_data[name] = pd.DataFrame()


# --- 5. Identifying Common Patterns ---

# Visual Inspection: Plotting normalized prices (re-indexed to 100)
print("\n--- Visualizing Normalized Price Paths (Indexed to 100) ---")
for name, df in normalized_prices_to_100.items():
    if not df.empty:
        plt.figure(figsize=(14, 7))
        for ticker in df.columns:
            plt.plot(df.index, df[ticker], label=ticker, alpha=0.7)
        plt.title(f'Normalized Price Paths (Indexed to 100) - {name} Frequency (Using Close Prices)')
        plt.xlabel('Date')
        plt.ylabel('Normalized Price (Start = 100)')
        plt.legend(loc='upper left', bbox_to_anchor=(1,1))
        plt.grid(True)
        plt.tight_layout()
        plt.show()
    else:
        print(f"No data to plot for normalized prices at {name} frequency.")

# Visual Inspection: Plotting returns
print("\n--- Visualizing Returns ---")
for name, df in returns_data.items():
    if not df.empty:
        plt.figure(figsize=(14, 7))
        for ticker in df.columns:
            plt.plot(df.index, df[ticker], label=ticker, alpha=0.7)
        if len(df.columns) > 1:
            sector_average_return = df.mean(axis=1)
            plt.plot(sector_average_return.index, sector_average_return, label='Sector Average', color='black', linewidth=2, linestyle='--')

        plt.title(f'Returns - {name} Frequency (Calculated from Close Prices)')
        plt.xlabel('Date')
        plt.ylabel('Percentage Return')
        plt.legend(loc='upper left', bbox_to_anchor=(1,1))
        plt.grid(True)
        plt.tight_layout()
        plt.show()
    else:
        print(f"No data to plot for returns at {name} frequency.")


# Correlation Analysis (on returns)
print("\n--- Correlation Analysis of Returns ---")
correlation_matrices = {}
for name, returns_df in returns_data.items():
    if not returns_df.empty and len(returns_df.columns) > 1:
        corr_matrix = returns_df.corr()
        correlation_matrices[name] = corr_matrix
        print(f"\nCorrelation Matrix - {name} Returns (from Close Prices):")
        print(corr_matrix)

        plt.figure(figsize=(10, 8))
        sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)
        plt.title(f'Correlation Matrix of {name} Returns (from Close Prices)')
        plt.tight_layout()
        plt.show()
    elif returns_df.empty:
         print(f"Skipping correlation for {name} returns as DataFrame is empty.")
    else:
        print(f"Skipping correlation for {name} returns as there's only one ticker or no data.")


# --- 6. If Pattern is Common Irrespective of Frequency ---
print("\n--- Analyzing Patterns Across Frequencies (Using Close Prices) ---")
print("Observations from visual plots and correlation matrices:")
print("1. Daily returns are typically more 'noisy' (volatile) than weekly, monthly, or yearly returns.")
print("2. Lower frequencies (Weekly, Monthly, Yearly) tend to smooth out short-term fluctuations, potentially revealing broader, more synchronized trends among tech stocks.")
print("3. Correlation strengths: Compare the average correlation values or the general pattern in heatmaps across frequencies.")
print("   - Note: Unadjusted 'Close' prices can introduce artificial jumps due to stock splits, which might affect correlation calculations, especially if splits occur at different times for different stocks.")

for name, corr_matrix in correlation_matrices.items():
    if not corr_matrix.empty:
        num_tickers = corr_matrix.shape[0]
        if num_tickers > 1:
            avg_corr = (corr_matrix.values.sum() - num_tickers) / (num_tickers * (num_tickers - 1))
            print(f"Average off-diagonal correlation for {name} returns: {avg_corr:.3f}")
        else:
            print(f"Cannot calculate average off-diagonal correlation for {name} returns (single ticker).")

# --- 7. If Pattern is Not Common, Can We Use Some Transformation? ---
print("\n--- Transformations to Potentially Reveal Commonality (Using Close Prices) ---")
print("The primary transformations we've already used are:")
print("  a) Resampling (Aggregating to Lower Frequencies)")
print("  b) Percentage Change (Returns)")
print("  c) Log Returns (Alternative to Pct_Change):")

if 'Daily' in resampled_prices and not resampled_prices['Daily'].empty:
    # *** Using resampled_prices which is now based on 'Close' prices ***
    daily_log_returns = np.log(resampled_prices['Daily'] / resampled_prices['Daily'].shift(1)).dropna()
    print("\nDaily Log Returns (from Close Prices - head):")
    print(daily_log_returns.head())

    if not daily_log_returns.empty and len(daily_log_returns.columns) > 1:
        daily_log_corr = daily_log_returns.corr()
        print("\nCorrelation Matrix - Daily Log Returns (from Close Prices):")
        print(daily_log_corr)
        plt.figure(figsize=(10, 8))
        sns.heatmap(daily_log_corr, annot=True, cmap='coolwarm', fmt=".2f")
        plt.title('Correlation Matrix of Daily Log Returns (from Close Prices)')
        plt.tight_layout()
        plt.show()

        num_tickers_log = daily_log_corr.shape[0]
        if num_tickers_log > 1:
            avg_log_corr = (daily_log_corr.values.sum() - num_tickers_log) / (num_tickers_log * (num_tickers_log - 1))
            print(f"Average off-diagonal correlation for Daily Log returns: {avg_log_corr:.3f}")
            if 'Daily' in correlation_matrices and not correlation_matrices['Daily'].empty:
                num_tickers_simple = correlation_matrices['Daily'].shape[0]
                if num_tickers_simple > 1:
                     avg_simple_corr = (correlation_matrices['Daily'].values.sum() - num_tickers_simple) / (num_tickers_simple * (num_tickers_simple - 1))
                     print(f"Average off-diagonal correlation for Daily Simple returns: {avg_simple_corr:.3f}")
    else:
        print("Could not compute or display correlation for daily log returns.")
else:
    print("Daily price data not available for log return calculation.")

print("\nFurther transformation ideas if patterns are still elusive:")
print("  d) Detrending")
print("  e) Volatility Normalization")
print("  f) Advanced Methods: PCA, DTW")

print("\n--- Summary of Analysis (Using Close Prices) ---")
print("1. Common Ground for Comparison:")
print("   - Percentage returns (from Close prices).")
print("   - Normalizing Close prices to a common starting point (e.g., 100).")
print("   - Z-score standardization of returns (from Close prices).")
print("   - BE AWARE: Stock splits in 'Close' prices can create artificial jumps, affecting these comparisons.")

print("\n2. Common Patterns Irrespective of Frequency:")
print("   - Similar observations as with 'Adj Close' are expected regarding general co-movement, but specific magnitudes of correlations or visual patterns might differ due to the lack of adjustment for splits/dividends.")

print("\n3. Transformations for Commonality:")
print("   - Resampling and Log returns remain key transformations.")
print("   - The impact of using 'Close' vs 'Adj Close' is most pronounced in the raw price series and any direct calculations from it (like pct_change).")

print("\nFinal Thoughts:")
print("  - Analysis proceeds similarly, but interpret results cautiously, knowing that 'Close' prices do not account for splits or dividends, which can introduce non-market-driven price changes.")
```

```python
########### HOURLY ############
# --- 1. Ticker List (Same as before) ---
tech_tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'NVDA', 'META', 'TSM', 'AVGO', 'ORCL', 'CSCO'] # Using 10 for brevity
# tech_tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'NVDA', 'META', 'TSM', 'AVGO', 'ORCL', 'CSCO', 'INTC', 'QCOM', 'AMD', 'CRM', 'ADBE']


print(f"Selected Tech Tickers: {tech_tickers}")

# --- 2. Define Time Period (Approx. 22 market days) ---
# yfinance intraday data is typically available for the last 60 days for 1h interval.
# We'll aim for roughly the last month of trading.
end_date = datetime.now()
# Go back a bit more than 22 business days to ensure we capture enough data after filtering
# (e.g., 45 calendar days should be safe to get ~30 business days)
start_date = end_date - timedelta(days=45)

print(f"Attempting to fetch data from {start_date.strftime('%Y-%m-%d')} to {end_date.strftime('%Y-%m-%d')}")

# --- 3. Download Hourly Data ---
all_hourly_data = {}
valid_tickers = []

for ticker in tech_tickers:
    try:
        # For '1h' interval, yfinance typically provides data for the last 730 days if period is not specified.
        # However, to be safe and specific, we use start/end. Max period for 1h is often 730d.
        # For very recent data, yfinance might have limitations on how far back 1h data goes.
        # Let's try to get a bit more than needed and then slice.
        # Max period for 1h interval is 730 days. For shorter periods like a month, start/end is fine.
        # yfinance often returns data for the last 60 *calendar* days for 1h interval if no explicit period is given.
        # Using start/end is more reliable for specific ranges.
        data = yf.download(ticker, start=start_date, end=end_date, interval="1h", progress=False)
        if not data.empty and 'Close' in data.columns:
            all_hourly_data[ticker] = data[['Close']]
            valid_tickers.append(ticker)
            print(f"Successfully downloaded hourly data for {ticker}")
        else:
            print(f"No hourly data found for {ticker} in the specified range.")
    except Exception as e:
        print(f"Could not download hourly data for {ticker}: {e}")

if not valid_tickers:
    print("No data downloaded for any ticker. Exiting.")
    exit()

tech_tickers = valid_tickers # Update ticker list to only those with data

# Combine into a single DataFrame
hourly_close_df = pd.concat(all_hourly_data, axis=1)
hourly_close_df.columns = hourly_close_df.columns.droplevel(1) # Remove 'Close' from multi-index
hourly_close_df = hourly_close_df.dropna(how='all') # Drop rows where all tickers are NaN

# Filter for typical US market hours (e.g., 9:30 AM to 4:00 PM ET)
# yfinance data is usually timezone-aware (often ET for US stocks)
# Let's assume the data is in ET. If not, tz_localize and tz_convert might be needed.
if hourly_close_df.index.tz is None:
    print("Warning: Timezone not set on index. Assuming ET. This might be incorrect.")
    hourly_close_df = hourly_close_df.tz_localize('America/New_York', ambiguous='NaT', nonexistent='NaT')
else:
    hourly_close_df = hourly_close_df.tz_convert('America/New_York')

hourly_close_df = hourly_close_df.between_time('09:30', '16:00') # Includes 9:30, 10:30 ... 15:30 (as Close of the hour bar)

# Ensure we have roughly 22 unique market days of data
unique_days = hourly_close_df.index.normalize().unique()
if len(unique_days) > 22:
    # Take the most recent 22 unique days
    start_of_22_days = unique_days[-22]
    hourly_close_df = hourly_close_df[hourly_close_df.index.normalize() >= start_of_22_days]
elif len(unique_days) < 5: # Arbitrary threshold for too little data
    print(f"Too few unique market days ({len(unique_days)}) with hourly data after filtering. Analysis might be unreliable.")
    if hourly_close_df.empty:
        print("DataFrame is empty after filtering. Exiting.")
        exit()

print(f"\nUsing data for {len(hourly_close_df.index.normalize().unique())} unique market days.")
print("Hourly Close Prices (head):")
print(hourly_close_df.head())

# --- 4. Normalize Data: Percentage Change from Market Open ---
# We need the opening price for each stock on each day.
# The first available price in our filtered data for each day can serve as a proxy for 'effective open'.
def calculate_pct_change_from_open(df):
    df_pct_change = df.copy()
    for day in df.index.normalize().unique():
        daily_data = df[df.index.normalize() == day]
        if not daily_data.empty:
            # Use the first valid price of the day as the open reference
            open_prices = daily_data.bfill().iloc[0] # Backfill to get first valid price if 9:30 is NaN
            for ticker in df.columns:
                if ticker in open_prices and pd.notna(open_prices[ticker]):
                    df_pct_change.loc[daily_data.index, ticker] = \
                        (daily_data[ticker] / open_prices[ticker] - 1) * 100
                else: # If no open price for this ticker on this day (should be rare after bfill)
                    df_pct_change.loc[daily_data.index, ticker] = np.nan
    return df_pct_change

normalized_hourly_df = calculate_pct_change_from_open(hourly_close_df)
normalized_hourly_df = normalized_hourly_df.dropna(how='all', axis=0) # Drop rows that became all NaN

if normalized_hourly_df.empty:
    print("Normalized DataFrame is empty. Cannot proceed with visualization. Check data or normalization logic.")
    exit()

print("\nNormalized Hourly Prices (% Change from Open - head):")
print(normalized_hourly_df.head())


# --- 5. Visualization ---

# Plot 1: Average Intraday Profile for All Tickers
# This shows the "common ground" or average hourly trend.
print("\nPlotting Average Intraday Profile for All Tickers...")
average_intraday_profile = normalized_hourly_df.groupby(normalized_hourly_df.index.time).mean()

plt.figure(figsize=(14, 7))
for ticker in average_intraday_profile.columns:
    plt.plot(average_intraday_profile.index.astype(str), average_intraday_profile[ticker], label=ticker, marker='o', linestyle='-')

plt.title(f'Average Intraday Profile (% Change from Open)\n(Averaged over last ~22 market days)')
plt.xlabel('Hour of Day (ET)')
plt.ylabel('% Change from Open')
plt.xticks(rotation=45)
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plot 2: Daily Intraday "Spaghetti Plot" for a Few Selected Tickers
# This shows if the trend remains the same over every hour across different market days for individual stocks.
print("\nPlotting Daily Intraday Profiles for Selected Tickers...")
tickers_to_plot_daily = tech_tickers[:3] # Plot for the first 3 tickers

for ticker in tickers_to_plot_daily:
    if ticker not in normalized_hourly_df.columns:
        print(f"Skipping daily plot for {ticker} as it's not in the normalized data.")
        continue

    plt.figure(figsize=(14, 7))
    # Pivot data to have days as columns (or iterate through days)
    # For plotting, iterating through days is straightforward
    unique_days_in_data = normalized_hourly_df.index.normalize().unique()

    for day in unique_days_in_data:
        daily_ticker_data = normalized_hourly_df.loc[normalized_hourly_df.index.normalize() == day, ticker]
        if not daily_ticker_data.empty:
            plt.plot(daily_ticker_data.index.time.astype(str), daily_ticker_data.values, marker='.', linestyle='-', alpha=0.5, label=day.strftime('%Y-%m-%d') if day == unique_days_in_data[0] else "_nolegend_") # Only label one day for legend clarity

    plt.title(f'Daily Intraday Profiles for {ticker} (% Change from Open)')
    plt.xlabel('Hour of Day (ET)')
    plt.ylabel('% Change from Open')
    plt.xticks(rotation=45)
    # plt.legend(title="Day", loc='upper left', bbox_to_anchor=(1,1)) # Can be too cluttered
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# Plot 3: Heatmap of Average Intraday Profiles
# This can also show commonality in shape.
print("\nPlotting Heatmap of Average Intraday Profiles...")
plt.figure(figsize=(10, 8))
sns.heatmap(average_intraday_profile.T, annot=True, cmap="RdYlGn", center=0, fmt=".2f") # Transpose for tickers on Y-axis
plt.title('Heatmap of Average Hourly % Change from Open')
plt.xlabel('Hour of Day (ET)')
plt.ylabel('Ticker')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

print("\nAnalysis Complete.")
print("Interpretation Notes:")
print("1. 'Average Intraday Profile': Look for similar shapes across tickers. Do they all tend to dip or rise at certain hours?")
print("2. 'Daily Intraday Profiles': For a single stock, see how consistent its hourly pattern is from one day to the next.")
print("   - High consistency = predictable intraday behavior.")
print("   - Low consistency = intraday behavior varies greatly day-to-day.")
print("3. 'Heatmap': Provides a compact view of the average profiles. Green might mean positive avg % change, Red negative.")
print("Common ground/trend: If many tickers show similar shapes in Plot 1 or similar color patterns in Plot 3 for specific hours, that indicates a common hourly trend.")
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import timedelta, datetime

# --- 1. Parameters ---
ticker_symbol = 'MSFT' # Correct ticker for Microsoft (NASDAQ)
# If you were analyzing an actual Indian stock, use its .NS or .BO ticker.
# For example: 'RELIANCE.NS'

num_days_for_weekly_profile = 5
print(f"Analyzing Ticker: {ticker_symbol}")

# --- 2. Get Hourly Data ---
end_date = datetime.now()
start_date = end_date - timedelta(days=20) # Fetch ~20 calendar days

try:
    hourly_data_df = yf.download(ticker_symbol, start=start_date, end=end_date, interval="1h", progress=False)
    if hourly_data_df.empty:
        raise ValueError(f"No data downloaded for {ticker_symbol}. It might be an invalid ticker or no data for the period.")
    
    # Ensure 'Close' column exists
    if 'Close' not in hourly_data_df.columns:
        raise ValueError(f"'Close' column not found in downloaded data for {ticker_symbol}. Columns: {hourly_data_df.columns}")
        
    # We expect 'Close' to be a direct column for a single ticker download
    close_prices_series = hourly_data_df['Close'].copy().dropna() # Explicitly take 'Close' as Series and copy
    
    if close_prices_series.empty:
        raise ValueError(f"Close price data is empty after dropna for {ticker_symbol}.")
        
    print(f"Downloaded {len(close_prices_series)} hourly records for {ticker_symbol}.")

except Exception as e:
    print(f"Error during data download or initial processing for {ticker_symbol}: {e}"); exit()

# Basic check for sufficient data length
if len(close_prices_series) < (num_days_for_weekly_profile * 6): # Approx 6 trading hours/day
    print(f"Not enough data points ({len(close_prices_series)}) for {num_days_for_weekly_profile} full days. Exiting."); exit()

# --- Helper: Normalize by Day's Open (More Robust) ---
def normalize_by_open(data_for_day_slice):
    """
    Normalizes hourly prices as % change from that day's open price.
    Accepts a pandas Series or a single-column pandas DataFrame.
    """
    series_to_normalize = None
    input_name = None

    if isinstance(data_for_day_slice, pd.DataFrame):
        if len(data_for_day_slice.columns) == 1:
            series_to_normalize = data_for_day_slice.iloc[:, 0].copy() # Get the first (and only) column as a Series
            input_name = series_to_normalize.name # Name of the column
        else:
            raise ValueError("normalize_by_open received a DataFrame with multiple columns.")
    elif isinstance(data_for_day_slice, pd.Series):
        series_to_normalize = data_for_day_slice.copy()
        input_name = series_to_normalize.name
    else:
        raise TypeError("Input to normalize_by_open must be a pandas Series or DataFrame.")

    if series_to_normalize.empty:
        return pd.Series(dtype=float, index=data_for_day_slice.index, name=input_name)

    # Initialize result Series
    normalized_result = pd.Series(index=series_to_normalize.index, dtype=float, name=input_name)
    open_price = series_to_normalize.bfill().iloc[0] # First valid price of the day

    if pd.notna(open_price) and open_price > 0:
        normalized_result = (series_to_normalize / open_price - 1) * 100
    else:
        normalized_result[:] = np.nan # Fill with NaNs if open price is invalid

    return normalized_result

# --- 3. Process 5-Day Average Hourly Profile ---
unique_days = close_prices_series.index.normalize().unique()
if len(unique_days) < num_days_for_weekly_profile:
    print(f"Need {num_days_for_weekly_profile} unique days, got {len(unique_days)}. Exiting."); exit()

days_for_profile = unique_days[-num_days_for_weekly_profile:]
print(f"Using days for profile: {days_for_profile.strftime('%Y-%m-%d').tolist()}")

normalized_profiles_list = []
for day_date in days_for_profile:
    # Get the slice of the original Series for the current day
    daily_close_data = close_prices_series[close_prices_series.index.normalize() == day_date]
    
    if not daily_close_data.empty:
        norm_profile = normalize_by_open(daily_close_data) # Pass the Series slice
        norm_profile.index = norm_profile.index.time # Use time for grouping
        normalized_profiles_list.append(norm_profile)

if not normalized_profiles_list: print("No data to process for weekly profile. Exiting."); exit()

avg_5day_profile = pd.concat(normalized_profiles_list, axis=1).mean(axis=1)
avg_5day_profile.name = f'Avg {num_days_for_weekly_profile}-Day Hourly Profile'

# --- 4. Process Single Day Hourly Profile (Most recent from the 5 days) ---
single_day_date_val = days_for_profile[-1]
# Get the slice of the original Series for the single day
single_day_data_slice = close_prices_series[close_prices_series.index.normalize() == single_day_date_val]

if single_day_data_slice.empty: print(f"No data for single day {single_day_date_val}. Exiting."); exit()

single_day_profile = normalize_by_open(single_day_data_slice) # Pass the Series slice
single_day_profile.index = single_day_profile.index.time
single_day_profile.name = f'Single Day ({single_day_date_val.strftime("%b %d")}) Profile'

# --- 5. Plot ---
plt.figure(figsize=(12, 6))
if not avg_5day_profile.empty:
    plt.plot(avg_5day_profile.index.astype(str), avg_5day_profile.values, label=avg_5day_profile.name, marker='o', lw=2)
if not single_day_profile.empty:
    plt.plot(single_day_profile.index.astype(str), single_day_profile.values, label=single_day_profile.name, marker='x', ls='--')

plt.title(f'{ticker_symbol}: Single Day vs. Avg {num_days_for_weekly_profile}-Day Hourly Trend (% Change from Open)')
plt.xlabel('Hour of Day (Market Timezone)')
plt.ylabel('% Change from Open')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, linestyle=':', alpha=0.7)
plt.axhline(0, color='black', lw=0.5, ls=':')
plt.tight_layout()
plt.show()

print(f"\nCompare the '{single_day_profile.name}' line with the '{avg_5day_profile.name}' line.")
```

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import timedelta, datetime

# --- 1. Parameters ---
ticker_symbol = 'MSFT' # Using a valid NASDAQ ticker
num_days_to_plot = 5
print(f"Analyzing Ticker: {ticker_symbol} for {num_days_to_plot} days.")

# --- 2. Get Hourly Data ---
end_date = datetime.now()
start_date = end_date - timedelta(days=20)

try:
    hourly_data_df = yf.download(ticker_symbol, start=start_date, end=end_date, interval="1h", progress=False)
    if hourly_data_df.empty: raise ValueError(f"No data for {ticker_symbol}.")
    if 'Close' not in hourly_data_df.columns: raise ValueError("'Close' column missing.")
    
    close_prices_series = hourly_data_df['Close'].copy().dropna()
    if close_prices_series.empty: raise ValueError("Close price data empty after dropna.")
    print(f"Downloaded {len(close_prices_series)} hourly records for {ticker_symbol}.")
except Exception as e:
    print(f"Error during data download/initial processing: {e}"); exit()

if len(close_prices_series.index.normalize().unique()) < num_days_to_plot:
    print(f"Not enough unique days ({len(close_prices_series.index.normalize().unique())}) for {num_days_to_plot}-day plot. Exiting."); exit()

# --- Helper: Normalize by Day's Open (Corrected) ---
def normalize_by_open(daily_series): # daily_series is expected to be a pandas Series
    if daily_series.empty:
        return pd.Series(dtype=float, name=daily_series.name if hasattr(daily_series, 'name') else None)

    # Ensure open_price is a scalar
    # .bfill() fills NaNs from next valid observation, .iloc[0] gets the first element.
    # This should be a scalar if daily_series is a Series.
    open_price_scalar = daily_series.bfill().iloc[0]

    # Explicitly check if it's still a Series (should not happen if daily_series is a Series)
    if isinstance(open_price_scalar, pd.Series):
        if not open_price_scalar.empty:
            open_price_scalar = open_price_scalar.iloc[0] # Try to get scalar again
        else:
            open_price_scalar = np.nan # If it was an empty series, treat as NaN

    if pd.notna(open_price_scalar) and open_price_scalar > 0:
        return (daily_series / open_price_scalar - 1) * 100
    else:
        # Return a Series of NaNs with the same index and name
        return pd.Series(np.nan, index=daily_series.index, name=daily_series.name if hasattr(daily_series, 'name') else None)

# --- 3. Process Data for Plotting ---
unique_days = close_prices_series.index.normalize().unique()
days_to_use = unique_days[-num_days_to_plot:]
print(f"Using data for days: {days_to_use.strftime('%Y-%m-%d').tolist()}")

all_normalized_daily_profiles = []
day1_profile = None

for i, day_date in enumerate(days_to_use):
    # daily_data_slice should be a Series here because close_prices_series is a Series
    daily_data_slice = close_prices_series[close_prices_series.index.normalize() == day_date]
    
    if not daily_data_slice.empty:
        normalized_profile = normalize_by_open(daily_data_slice) # Pass the Series
        
        if i == 0: # Store Day 1 profile
            day1_profile = normalized_profile.copy()
            if not day1_profile.empty: # Check if normalization resulted in empty series (e.g. all NaN)
                day1_profile.index = day1_profile.index.time
                day1_profile.name = f'Day 1 ({day_date.strftime("%b %d")}) Hourly Profile'
            else: # Handle case where day1_profile might be all NaNs and thus empty after potential dropna in normalize
                day1_profile = pd.Series(dtype=float, name=f'Day 1 ({day_date.strftime("%b %d")}) Hourly Profile (No Data)')


        if not normalized_profile.empty:
            all_normalized_daily_profiles.append(normalized_profile.reset_index(drop=True))

if day1_profile is None or day1_profile.empty: # Check if day1_profile is effectively empty
    print("Could not process Day 1 profile (it might be all NaNs or empty). Exiting.");
    # Provide a placeholder if it's None or truly empty to avoid errors in plotting
    if day1_profile is None: day1_profile = pd.Series(dtype=float, name='Day 1 (No Data)')


if not all_normalized_daily_profiles:
    print("Could not process any daily profiles for the extended plot. Exiting."); exit()

extended_5day_profile = pd.concat(all_normalized_daily_profiles, ignore_index=True)
extended_5day_profile.name = f'Extended {num_days_to_plot}-Day Hourly Profile'

# --- 4. Plot ---
fig, ax = plt.subplots(2, 1, figsize=(15, 10), sharey=True)

# Plot 1: Day 1 Hourly Profile
ax[0].plot(day1_profile.index.astype(str) if not day1_profile.empty else [],
           day1_profile.values if not day1_profile.empty else [],
           label=day1_profile.name, marker='o', linestyle='-')
ax[0].set_title(f'{ticker_symbol}: {day1_profile.name}')
ax[0].set_xlabel('Hour of Day 1 (Market Timezone)')
ax[0].set_ylabel('% Change from Open')
ax[0].tick_params(axis='x', rotation=45)
ax[0].grid(True, linestyle=':', alpha=0.7)
ax[0].axhline(0, color='black', lw=0.5, ls=':')
ax[0].legend()

# Plot 2: Extended 5-Day Hourly Profile
if not extended_5day_profile.empty:
    ax[1].plot(extended_5day_profile.index, extended_5day_profile.values,
               label=extended_5day_profile.name, linestyle='-')
    ax[1].set_title(f'{ticker_symbol}: {extended_5day_profile.name} (Concatenated)')
    ax[1].set_xlabel(f'Trading Hours (Concatenated over {num_days_to_plot} Days)')
    ax[1].set_ylabel('% Change from Open')
    ax[1].grid(True, linestyle=':', alpha=0.7)
    ax[1].axhline(0, color='black', lw=0.5, ls=':')
    ax[1].legend()
    
    hours_per_day_list = [len(prof) for prof in all_normalized_daily_profiles if not prof.empty]
    current_x_pos = 0
    for day_idx in range(len(hours_per_day_list) - 1):
        current_x_pos += hours_per_day_list[day_idx]
        ax[1].axvline(x=current_x_pos - 0.5, color='gray', linestyle='--', alpha=0.8)

plt.tight_layout()
plt.show()

print("\nInterpretation Guide:")
print("- Top Plot: Shows the hourly % change from open for the first day of the selected period.")
print("- Bottom Plot: Shows the hourly % change from each day's open, concatenated sequentially over the 5 days.")
print("  (Vertical dashed lines separate the approximate days).")
print("- Compare the shape of the Day 1 plot to the pattern within the first segment of the bottom plot.")
print("- Then, observe if the patterns seen in subsequent segments of the bottom plot (Day 2, Day 3, etc.)")
print("  resemble or deviate from the Day 1 pattern.")
```
