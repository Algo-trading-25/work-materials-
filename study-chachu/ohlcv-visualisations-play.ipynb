```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import sklearn as skl
import seaborn as sb
```

```python
df = pd.read_csv('/Users/apple/Downloads/test_msft_goog_aapl (1).csv')
```

```python
x = df[['Close']]
y = df[['Volume']]
```

```python
plt.figure(figsize=(8,4))
plt.scatter(x, y, alpha=0.7, c='steelblue', s=50)
plt.xlabel('close')
plt.ylabel('vol')
plt.title('scatterplot of close to volume')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
data = df.select_dtypes(include=['number'])

corr_matrix = data.corr()
plt.figure(figsize=(8,4))
sb.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0 , square=True, linewidths=0.5, cbar_kws={"shrink":.8})
plt.title('corr heatmap')
plt.tight_layout()
plt.show()
```

```python

# Ensure 'Date' is datetime
df['Date'] = pd.to_datetime(df['Date'])

dates = pd.date_range(start='2020-04-16', periods=22, freq='D')

filtered_df = df[df['Date'].isin(dates)].sort_values('Date')


values = filtered_df['Close'].values
dates_to_plot = filtered_df['Date'].values

plt.figure(figsize=(8,4))
plt.plot(dates_to_plot, values, linewidth=2, color='steelblue', alpha=0.8)
plt.fill_between(dates_to_plot, values, alpha=0.3, color='lightblue')

plt.title('values over time', fontsize=16, fontweight='bold')
plt.xlabel('date', fontsize=12)
plt.ylabel('value', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#why the line graph so weird???
#ask chachu
#also why the extra filtering in filtered_dates line? doesn't auto-select after periods is defined=22??
#why ensure date is datetime when the col itself is defined in datetime manner????
```
